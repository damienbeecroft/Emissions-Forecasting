```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
path_to_data = '/mmfs1/gscratch/amath/dob1998/emissions_forecasting/data/MRSUT/'
path_to_save = '/mmfs1/gscratch/amath/dob1998/emissions_forecasting/'
```

```python
sup1995 = pd.read_csv(path_to_data + 'MRSUT_1995/supply.csv',header = [0,1], index_col = [0,1], sep='\t')
use1995 = pd.read_csv(path_to_data + 'MRSUT_1995/use.csv',header = [0,1], index_col = [0,1], sep='\t')
```

```python
log_sup1995 = np.log10(sup1995 + 1)
max_sup = np.max(log_sup1995)
log_use1995 = np.log10(use1995 + 1)
max_use = np.max(log_use1995)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(log_sup1995,vmax=max_sup/10)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(log_use1995,vmax=max_use/10)
plt.colorbar()
plt.show()
```
