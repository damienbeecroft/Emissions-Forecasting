```python
import pandas as pd

# Define multi-index for rows
index = pd.MultiIndex.from_tuples([('North', 'Agriculture'), 
                                   ('North', 'Manufacturing'), 
                                   ('South', 'Agriculture'), 
                                   ('South', 'Manufacturing')],
                                   names=['Region', 'Industry'])

# Create DataFrame
df = pd.DataFrame({'Value': [100, 200, 150, 250]}, index=index)

# Get unique names in the multi-index
unique_row_names = df.index.levels
for i, level in enumerate(df.index.names):
    print(f"Unique names in level '{level}':", unique_row_names[i].tolist())
```
